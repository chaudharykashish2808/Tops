```python
# (58) Write a Python program to combine values in python list of dictionaries.Sample data: [{'item': 'item1', 'amount': 400}, {'item': 'item2', 'amount':300}, o {'item': 'item1', 'amount': 750}] 
from collections import Counter

data = [
    {'item': 'item1', 'amount': 400},
    {'item': 'item2', 'amount': 300},
    {'item': 'item1', 'amount': 750}
]

c = Counter()
for d in data:
    c[d['item']] += d['amount']

print(c)
```

```python
# (63)Write a Python function to check whether a number is perfect or not.
    def perfect(n):
        s = 0
    for i in range(1, n):
        if n % i == 0:
            s += i
    return s == n

print(perfect(6))
```

```python
# (87)  When is the finally block executed? 
#The finally block is always executed, no matter what happens in the program.

#It runs:

#✔ If no exception occurs
#✔ If an exception occurs
#✔ Even if there is a return statement
try:
    print("Inside try block")
    x = 10 / 2
except:
    print("Error occurred")
finally:
    print("Finally block executed")
```

```python
# (89) How Do You Handle Exceptions with Try/Except/Finally in Python? 

#try → risky code
#except → handle error
#finally → always runs
#  Example:
try:
    num = int(input("Enter a number: "))
    result = 10 / num
    print("Result:", result)

except ZeroDivisionError:
    print("Cannot divide by zero!")

except ValueError:
    print("Invalid input!")

finally:
    print("Program finished")
```

```python
# (90)  Write python program that user to enter only odd numbers, else will raise an exception.

try:
    num = int(input("Enter an odd number: "))

    if num % 2 == 0:
        raise ValueError("Even number entered!")

    print("You entered a valid odd number:", num)

except ValueError as e:
    print("Error:", e)

finally:
    print("Execution completed")
```
